# The prepare-and-measure scenario is used to obtain the two-time correlation function.

### We prepare the state with probability larger than 0.01 from pick_t2

### Each circuit below represents the "preparing" state at t2 and evolve to t3

## NOTE: We remain the source code for obtaining the pt2 (deciding the state which we used in this source code) and pt3 in 6-qubit-prepare-and-measure pt2-pt3

# Without Flip

```python
q=QuantumRegister(7)
c=ClassicalRegister(6)
qc=QuantumCircuit(q,c)
qc.cx(q[5],q[6])
qc.cx(q[4],q[5])
qc.cx(q[3],q[4])
qc.cx(q[2],q[3])
qc.cx(q[1],q[2])
qc.h(q[1])
# the routine above is to make the system go back to its initial state
qc.measure(q[1],c[0])
qc.measure(q[2],c[1])
qc.measure(q[3],c[2])
qc.measure(q[4],c[3])
qc.measure(q[5],c[4])
qc.measure(q[6],c[5])
```

# One Flip

```python
def Pt2t3_1(a):
    q=QuantumRegister(7)
    c=ClassicalRegister(6)
    qc=QuantumCircuit(q,c)
    if a==0 :
        #|000001>
        qc.x(q[1])
    if a==1 :
        #|000010>
        qc.x(q[2])
    if a==2 :
        #|000100>
        qc.x(q[3])
    if a==3 :
        #|001000>
        qc.x(q[4])
    if a==4 :
        #|010000>
        qc.x(q[5])
    if a==5 :
        #|100000>
        qc.x(q[6])
        
    qc.cx(q[5],q[6])
    qc.cx(q[4],q[5])
    qc.cx(q[3],q[4])
    qc.cx(q[2],q[3])
    qc.cx(q[1],q[2])
    qc.h(q[1])
    # the routine above is to make the system go back to its initial state
    qc.measure(q[1],c[0])
    qc.measure(q[2],c[1])
    qc.measure(q[3],c[2])
    qc.measure(q[4],c[3])
    qc.measure(q[5],c[4])
    qc.measure(q[6],c[5])
    
    return qc
```
